```python
indicess=['dB_tab_exportTimeDocs', 'dB_tab_importTimeDocs',
       'dB_tab_exportTimeCompliance', 'dB_tab_importTimeCompliance',
       'dB_tab_exportCostDocs', 'dB_tab_importCostDocs',
       'dB_tab_exportCostCompliance', 'dB_tab_importCostCompliance']
```

```python
def train_pred(train,pred_length=1):
# #    print('raw train',type(train),train)
    
    #全为缺失值，就填充为-1
    if train.isnull().sum()/len(train)==1:
        train=train.fillna(-1)
    else:
        #存在部分缺失值时，中间缺失值用前面非缺失值填充，前面如果全是缺失值就删除
        train=train.fillna(method='ffill')
        train=train.dropna().astype(float)
#     print('new train',type(train),train,'\n***\n')
    train=train.values
    
    
#     print('train ',train,'\n***\n')
    fit1 =Holt(train,
                damped=True,
                # exponential=True
                ).fit(
        # smoothing_level = 0.95,
        smoothing_slope = 0.05,
        # damping_slope=0.99,
        optimized=True
       )#0.7,0.1
    forecast_res = fit1.forecast(pred_length)
#     print('forecast_res',forecast_res)
    return forecast_res
#     return train.iloc[:3]
    

def sort_time(df,pred_length):
    df=df.sort_values('year_new', ascending=True)#.apply(lambda col:print(col,'\n**\n'),axis=0)#
    print('new df',df)
    df_pred=df[indicess].apply(lambda col:train_pred(col,pred_length=pred_length) ,axis=0)
    #添加经济体和预测的年份
    df_pred['economyName']=df['economyName'].iloc[:pred_length]
    df_pred['year_new']=range(df['year_new'].iloc[-1]+1,pred_length+df['year_new'].iloc[-1]+1)
    print('df_pred',df_pred)
        
    return df_pred
```

```python
dff_pred_res=dff.groupby('economyName').apply(lambda df:sort_time(df,pred_length=3)).reset_index(drop=True)
```
